```python
"""
Python Script for Generating White Matter Tract Dice Score Figures
This script creates publication-quality grouped bar charts for research manuscripts.
"""

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

# ============================================================
# Configuration: Set Publication-Quality Parameters
# ============================================================

rcParams['font.family'] = 'sans-serif'
rcParams['font.size'] = 10
rcParams['axes.linewidth'] = 1.0
rcParams['xtick.major.width'] = 0.8
rcParams['ytick.major.width'] = 0.8

# ============================================================
# Load Data
# ============================================================

# Read the Excel file containing dice scores
df = pd.read_excel('/content/full_tract_table.xlsx')

# Extract data
tracts = df['Tract'].values
model1 = df['fODF Peaks'].values
model2 = df['Rank-3 Approx.'].values
model3 = df['fODF Tensors'].values

# ============================================================
# FIGURE 1: Single Plot with All 72 Tracts
# ============================================================

def create_single_plot():
    """Create a single grouped bar chart with all 72 tracts"""

    fig, ax = plt.subplots(figsize=(16, 10), dpi=300)

    # Define bar positions and width
    x = np.arange(len(tracts))
    width = 0.27

    # Define colors (colorblind-friendly palette)
    colors = ['#2E86AB', '#A23B72', '#F18F01']  # Blue, Purple-pink, Orange

    # Create grouped bars
    bars1 = ax.bar(x - width, model1, width, label='fODF Peaks',
                   color=colors[0], edgecolor='white', linewidth=0.5, alpha=0.9)
    bars2 = ax.bar(x, model2, width, label='Rank-3 Approx.',
                   color=colors[1], edgecolor='white', linewidth=0.5, alpha=0.9)
    bars3 = ax.bar(x + width, model3, width, label='fODF Tensors',
                   color=colors[2], edgecolor='white', linewidth=0.5, alpha=0.9)

    # Customize axes
    ax.set_xlabel('White Matter Tracts', fontsize=12, fontweight='bold', labelpad=10)
    ax.set_ylabel('Dice Score', fontsize=12, fontweight='bold', labelpad=10)
    ax.set_title('Dice Score Comparison Across 72 White Matter Tracts',
                 fontsize=14, fontweight='bold', pad=20)

    # Set x-axis ticks and labels
    ax.set_xticks(x)
    ax.set_xticklabels(tracts, rotation=90, ha='right', fontsize=7)

    # Set y-axis range and ticks
    ax.set_ylim([0.60, 0.95])
    ax.set_yticks(np.arange(0.60, 0.96, 0.05))

    # Add gridlines
    ax.yaxis.grid(True, linestyle='--', alpha=0.3, linewidth=0.5)
    ax.set_axisbelow(True)

    # Add legend
    ax.legend(loc='lower right', frameon=True, fancybox=True, shadow=True,
              fontsize=10, framealpha=0.95)

    # Remove top and right spines
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    # Adjust layout
    plt.tight_layout()

    # Save figure in multiple formats
    plt.savefig('white_matter_dice_scores.png', dpi=300, bbox_inches='tight')
    plt.savefig('white_matter_dice_scores.pdf', dpi=300, bbox_inches='tight')
    plt.savefig('white_matter_dice_scores.eps', dpi=300, bbox_inches='tight')  # For some journals

    print("Single plot created: white_matter_dice_scores.pdf")
    plt.close()


# ============================================================
# FIGURE 2: Four-Panel Plot (18 Tracts per Panel)
# ============================================================

def create_four_panel_plot():
    """Create a 2x2 grid of subplots, each showing 18 tracts"""

    # Configuration
    n_per_subplot = 18
    n_subplots = 4

    # Create figure with 4 subplots (2x2 grid)
    fig, axes = plt.subplots(2, 2, figsize=(18, 12), dpi=300)
    axes = axes.flatten()

    # Define colors
    colors = ['#2E86AB', '#A23B72', '#F18F01']
    width = 0.27

    for idx in range(n_subplots):
        ax = axes[idx]

        # Get tract indices for this subplot
        start_idx = idx * n_per_subplot
        end_idx = min((idx + 1) * n_per_subplot, len(tracts))

        # Extract data for this subplot
        tracts_sub = tracts[start_idx:end_idx]
        model1_sub = model1[start_idx:end_idx]
        model2_sub = model2[start_idx:end_idx]
        model3_sub = model3[start_idx:end_idx]

        # Set up bar positions
        x = np.arange(len(tracts_sub))

        # Create grouped bars
        bars1 = ax.bar(x - width, model1_sub, width, label='fODF Peaks',
                       color=colors[0], edgecolor='white', linewidth=0.5, alpha=0.9)
        bars2 = ax.bar(x, model2_sub, width, label='Rank-3 Approx.',
                       color=colors[1], edgecolor='white', linewidth=0.5, alpha=0.9)
        bars3 = ax.bar(x + width, model3_sub, width, label='fODF Tensors',
                       color=colors[2], edgecolor='white', linewidth=0.5, alpha=0.9)

        # Customize axes
        ax.set_xlabel('White Matter Tracts', fontsize=11, fontweight='bold', labelpad=8)
        ax.set_ylabel('Dice Score', fontsize=11, fontweight='bold', labelpad=8)
        ax.set_title(f'Panel {chr(65+idx)}: Tracts {start_idx+1}-{end_idx}',
                     fontsize=12, fontweight='bold', pad=12)

        # Set x-axis ticks and labels
        ax.set_xticks(x)
        ax.set_xticklabels(tracts_sub, rotation=90, ha='right', fontsize=8)

        # Set y-axis (consistent range across all panels)
        ax.set_ylim([0.60, 0.95])
        ax.set_yticks(np.arange(0.60, 0.96, 0.05))

        # Add gridlines
        ax.yaxis.grid(True, linestyle='--', alpha=0.3, linewidth=0.5)
        ax.set_axisbelow(True)

        # Add legend only to the first subplot
        if idx == 0:
            ax.legend(loc='lower right', frameon=True, fancybox=True, shadow=True,
                     fontsize=9, framealpha=0.95)

        # Remove top and right spines
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)

    # Add overall title
    fig.suptitle('Dice Score Comparison Across 72 White Matter Tracts',
                 fontsize=15, fontweight='bold', y=0.995)

    # Adjust layout
    plt.tight_layout(rect=[0, 0, 1, 0.99])

    # Save figure in multiple formats
    plt.savefig('white_matter_dice_scores_4panels.png', dpi=300, bbox_inches='tight')
    plt.savefig('white_matter_dice_scores_4panels.pdf', dpi=300, bbox_inches='tight')
    plt.savefig('white_matter_dice_scores_4panels.eps', dpi=300, bbox_inches='tight')

    print("Four-panel plot created: white_matter_dice_scores_4panels.pdf")
    plt.close()


# ============================================================
# Print Summary Statistics
# ============================================================

def print_statistics():
    """Print summary statistics for all three models"""

    print("\n" + "="*70)
    print("SUMMARY STATISTICS")
    print("="*70)
    print(f"{'Model':<20} {'Mean':<12} {'Std Dev':<12} {'Min':<12} {'Max':<12}")
    print("-"*70)
    print(f"{'fODF Peaks':<20} {model1.mean():.6f}   {model1.std():.6f}   "
          f"{model1.min():.6f}   {model1.max():.6f}")
    print(f"{'Rank-3 Approx.':<20} {model2.mean():.6f}   {model2.std():.6f}   "
          f"{model2.min():.6f}   {model2.max():.6f}")
    print(f"{'fODF Tensors':<20} {model3.mean():.6f}   {model3.std():.6f}   "
          f"{model3.min():.6f}   {model3.max():.6f}")
    print("="*70)
    print()


# ============================================================
# Main Execution
# ============================================================

if __name__ == "__main__":
    print("Generating publication-quality figures...\n")

    # Create both figures
    create_single_plot()
    create_four_panel_plot()

    # Print statistics
    print_statistics()

    print("All figures generated successfully!")
    print("\nGenerated files:")
    print("  - white_matter_dice_scores.png/pdf/eps (single plot)")
    print("  - white_matter_dice_scores_4panels.png/pdf/eps (four-panel plot)")
```
